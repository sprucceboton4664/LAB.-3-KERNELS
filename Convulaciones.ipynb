```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def convulcion_media_h(imagen,tkernel):
    kernel= np.ones((1,tkernel), np.float32)/tkernel
    imagen_c= cv2.filter2D(imagen,-1,kernel)
    return imagen_c

img=cv2.imread('img1h.png',0)

img_fin= convulcion_media_h(img,15)
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.subplot(122)
plt.imshow(img_fin, cmap='gray')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def convulcion_media_h(imagen,tkernel):
    kernel= np.ones((1,tkernel), np.float32)/tkernel
    imagen_c= cv2.filter2D(imagen,-1,kernel)
    return imagen_c

img=cv2.imread('img1h.png')
img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_fin= convulcion_media_h(img,15)
img_fin_rgb=cv2.cvtColor(img_fin, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img_rgb)
plt.subplot(122)
plt.imshow(img_fin_rgb)
plt.show()
```
